```python
import torch

from domains.starDomain import Sphere, HyperCuboid
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
from scipy.interpolate import griddata
from ipywidgets import *
import numpy as np
%matplotlib widget
```

# Spheres

## 2D

```python
center2DSphere = torch.tensor([1.,1.])
radius2DSphere = torch.tensor(2.)
```

```python
sphere2DForPoints  = Sphere(dim= 2,center=center2DSphere, radius=radius2DSphere)
```

```python
sphere2DPointsCartesinan = torch.cat(sphere2DForPoints.generateCartesianRandomPointsFullDomain(3000), dim=1)
sphere2DPointsSpherical = torch.cat(sphere2DForPoints.generateSphericalRandomPointsFullDomain(3000), dim=1)
sphere2DPointsBoundary = torch.cat(sphere2DForPoints.generateSphericalRandomPointsOnBoundary(500), dim=1)
```

```python
figS2D,axS2D = plt.subplots(1,3)
figS2D.set_figwidth(12)
figS2D.set_figheight(4)
axS2D[0].set_title('Cartesian 2D Sphere')
axS2D[0].set_xlabel('X')
axS2D[0].set_ylabel('Y')

axS2D[1].set_title('Spherical 2D Sphere')
axS2D[1].set_xlabel('X')
axS2D[1].set_ylabel('Y')

axS2D[2].set_title('Boundary 2D Sphere')
axS2D[2].set_xlabel('X')
axS2D[2].set_ylabel('Y')

axS2D[0].scatter(sphere2DPointsCartesinan[:,0], sphere2DPointsCartesinan[:,1], s= 0.3)
axS2D[1].scatter(sphere2DPointsSpherical[:,0], sphere2DPointsSpherical[:,1], s= 0.3)
axS2D[2].scatter(sphere2DPointsBoundary[:,0], sphere2DPointsBoundary[:,1], s= 0.3)
plt.show()
```

## 3D

```python
center3DSphere = torch.tensor([1.,1.,1.])
radius3DSphere = torch.tensor(2.)
```

```python
sphere3DForPoints  = Sphere(dim= 3,center=center3DSphere, radius=radius3DSphere)
```

```python
sphere3DPointsCartesinan = torch.cat(sphere3DForPoints.generateCartesianRandomPointsFullDomain(200), dim=1)
sphere3DPointsSpherical = torch.cat(sphere3DForPoints.generateSphericalRandomPointsFullDomain(200), dim=1)
sphere3DPointsBoundary = torch.cat(sphere3DForPoints.generateSphericalRandomPointsOnBoundary(200), dim=1)
```

```python
numPointsPlot = 1000
x3DSpherCart = sphere3DPointsCartesinan[:numPointsPlot,0].view(-1).detach().numpy()
y3DSpherCart = sphere3DPointsCartesinan[:numPointsPlot,1].view(-1).detach().numpy()
z3DSpherCart = sphere3DPointsCartesinan[:numPointsPlot,2].view(-1).detach().numpy()

x3DSpherSpher = sphere3DPointsSpherical[:numPointsPlot,0].view(-1).detach().numpy()
y3DSpherSpher = sphere3DPointsSpherical[:numPointsPlot,1].view(-1).detach().numpy()
z3DSpherSpher = sphere3DPointsSpherical[:numPointsPlot,2].view(-1).detach().numpy()

x3DSpherBd = sphere3DPointsBoundary[:numPointsPlot,0].view(-1).detach().numpy()
y3DSpherBd = sphere3DPointsBoundary[:numPointsPlot,1].view(-1).detach().numpy()
z3DSpherBd = sphere3DPointsBoundary[:numPointsPlot,2].view(-1).detach().numpy()

figS3D = plt.figure(figsize=(9, 5))
# Create a 3D plot
axS3D1 = figS3D.add_subplot(131, projection='3d')
axS3D2 = figS3D.add_subplot(132, projection='3d')
axS3D3 = figS3D.add_subplot(133, projection='3d')

#set labels 
axS3D1.set_title('Cartesian 3D Sphere')
axS3D1.set_xlabel('X')
axS3D1.set_ylabel('Y')

axS3D2.set_title('Spherical 3D Sphere')
axS3D2.set_xlabel('X')
axS3D2.set_ylabel('Y')

axS3D3.set_title('Boundary 3D Sphere')
axS3D3.set_xlabel('X')
axS3D3.set_ylabel('Y')


# Plot the surface
axS3D1.scatter(x3DSpherCart, y3DSpherCart, z3DSpherCart , s = 0.5)
axS3D2.scatter(x3DSpherSpher, y3DSpherSpher, z3DSpherSpher, s = 0.5)
axS3D3.scatter(x3DSpherBd, y3DSpherBd, z3DSpherBd, s = 0.5)



# Show plot
plt.tight_layout()
plt.show()
```
